Line Plot: A line plot is a graph that displays data as a series of points connected by straight lines. It is useful for visualizing trends over time.

Scatter Plot: A scatter plot is a graph that displays data as a collection of points. It is useful for visualizing the relationship between two variables.

Bar Plot: A bar plot is a graph that displays data as a series of bars. It is useful for comparing data between different categories.

Histogram: A histogram is a graph that displays data as a series of bars, where the height of each bar represents the frequency of occurrence of a particular value.

Pie Chart: A pie chart is a graph that displays data as a circular chart divided into sectors, where the size of each sector represents the proportion of the data it represents. It is useful for showing the relative proportions of different categories.

Each point in the plot represents the value of the two variables, where one variable is plotted along the x-axis and the other along the y-axis. Scatter plots are useful for identifying trends and patterns in data and for visualizing the distribution of the data.

The subplot() function in matplotlib allows us to create multiple plots within a single figure. It takes three arguments: the number of rows of subplots, the number of columns of subplots, and the index of the current subplot (starting from 1 and incrementing from left to right and top to bottom).


```python
import numpy as np
import matplotlib.pyplot as plt

x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])

x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])

x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])

x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])

fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(8, 6))

axes[0, 0].plot(x1, y1)
axes[0, 0].set_title('Line 1')

axes[0, 1].plot(x2, y2)
axes[0, 1].set_title('Line 2')

axes[1, 0].plot(x3, y3)
axes[1, 0].set_title('Line 3')

axes[1, 1].plot(x4, y4)
axes[1, 1].set_title('Line 4')

fig.suptitle('Four Line Plots')

plt.show()
```

A box  plot, also known as a box-and-whisker plot, is a type of graph used to display the distribution of a set of continuous data. It is called a box plot because it is composed of a box that represents the interquartile range (IQR) of the data, and "whiskers" that extend from the box to represent the rest of the distribution.

The box in a box plot represents the IQR, which is the range between the 25th and 75th percentiles of the data. The line inside the box represents the median of the data. The whiskers extend from the box to show the range of the data, but they do not include outliers, which are usually shown as individual points.

Box plots are used to quickly summarize the distribution of a dataset and to identify potential outliers. They are particularly useful when comparing the distributions of multiple datasets, as they allow for easy visual comparison of the medians, IQRs, and ranges of the data. Box plots can also reveal asymmetry or skewness in the data, which is not always apparent from other types of plots.


```python

box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

data = [box1, box2]

fig, ax = plt.subplots()

ax.boxplot(data)

ax.set_xticklabels(['Box 1', 'Box 2'])
ax.set_ylabel('Values')

ax.set_title('Box Plot Example')

plt.show()

```
